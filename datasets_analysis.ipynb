```python
import numpy as np
from utils import datasets
```

```python
def analyze(dataset):
    imbalance = None
    if dataset == 'digits':
        imbalance = datasets.HIGH_IMBALANCE_AMOUNTS_DIGITS_DIFFERENT_PY
    elif dataset == 'domainnet':
        imbalance = datasets.IMBALANCE_AMOUNTS_DOMAINNET_DIFFERENT_PY
    x_train, x_test, y_train, y_test, classes = datasets.load(dataset, imbalance=imbalance)
    domains = x_train.keys()
    for domain in domains:
        print(domain, 'train', x_train[domain].shape)
        classes, counts = np.unique(y_train[domain], return_counts=True)
        print(classes, counts, counts / sum(counts))
        print(domain, 'test', x_test[domain].shape)
        classes, counts = np.unique(y_test[domain], return_counts=True)
        print(classes, counts, counts / sum(counts))
        print()
```

```python
analyze('digits')
```

```python
analyze('domainnet')
```

```python
analyze('covid')
```
